```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
csv_path = r"C:\Users\estel\OneDrive\Escritorio\NEUROCIENCIAS\QUINTO SEMESTRE\MODELOS\Neurociencias-2025-1\datasets\cancer_regression\cancer_reg.csv"
df_cancer = pd.read_csv(csv_path, sep=",")
df_cancer.columns.to_list()
```

```python
# usar método seaborn para sacar media y desviación estándar de la distribución
median_age = df_cancer[df_cancer['medianage'] < 100]
median_age.describe()
```

```python
sns.histplot(df_cancer['medianage'], kde=True)
plt.title('distribución medianage')
```

```python
sns.scatterplot(df_cancer['medianage'])
plt.title('scatter plot de medianage')
```

```python
filtrado = df_cancer[df_cancer['medianage'] < 100]
sns.scatterplot(x = 'incidencerate', y = 'medianage', data= filtrado)
plt.title('scatter plot de median age contra incidence rate')
```

```python
sns.regplot(x = 'incidencerate', y = 'medianage', data= filtrado)
plt.title('regresión lineal de median age contra incidence rate')
```

```python
slope, intercept = np.polyfit(df_cancer['incidencerate'], df_cancer['medianage'], 1)
slope, intercept
```
